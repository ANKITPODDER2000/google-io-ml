```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# 0 -> Apple 1 -> Orange , Apple -> High Weight , not bumpy skin
fruit_features = [[160 , 1] , [180 , 1] ,[165 , 0], [140 , 0] , [120 , 0] , [155 , 1]]
fruit_name     = [0 , 0 , 1 , 1 , 1 , 0]
class_names    = {0 : 'Apple' , 1 : 'Orange'}

from sklearn.tree import export_graphviz , DecisionTreeClassifier
import graphviz 

clf = DecisionTreeClassifier().fit(fruit_features , fruit_name)
#print(clf)
dt = export_graphviz(clf , out_file="fruit_tree.dot" , feature_names=['weight' , 'skin'] , class_names=['Apple' , 'Orange'] , filled=True)

with open("./fruit_tree.dot") as fl:
  content = fl.read()
graphviz.Source(content)
```

```
print("Fruit must be : Apple!")
print("Prediction is : ",class_names[clf.predict([[168,1]])[0]])
print("-----------------------------")
print("Fruit must be : Orange!")
print("Prediction is : ",class_names[clf.predict([[146,0]])[0]])
```
